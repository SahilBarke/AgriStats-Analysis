## Data Normalization – Part 1
---
The dataset currently contains **year-wise columns** for each metric, for example:

* `Area-2020-21`, `Area-2021-22`, …
* `Production-2020-21`, `Production-2021-22`, …
* `Yield-2020-21`, `Yield-2021-22`, …

To make the data **analysis-ready**, we will normalize it by:

1. **Melting the year-wise columns** into a single `Year` column.
2. Creating separate columns for `Area`, `Production`, and `Yield`.
3. Aggregating the values by **Crop, State, and Year**, so that each row represents a **unique Crop–State–Season combination**.

This normalization enables:

* Easier **time-series analysis** across seasons,
* Simpler **visualizations**, and
* More consistent **aggregations and calculations** by crop, state, or season.

```python
id_cols = ["Crop", "State", "Season"]

area_df = df.melt(
    id_vars=id_cols,
    value_vars=[c for c in df.columns if c.startswith("Area-")],
    var_name="Year",
    value_name="Area"
)
area_df["Year"] = area_df["Year"].str.replace("Area-", "")

prod_df = df.melt(
    id_vars=id_cols,
    value_vars=[c for c in df.columns if c.startswith("Production-")],
    var_name="Year",
    value_name="Production"
)
prod_df["Year"] = prod_df["Year"].str.replace("Production-", "")

yield_df = df.melt(
    id_vars=id_cols,
    value_vars=[c for c in df.columns if c.startswith("Yield-")],
    var_name="Year",
    value_name="Yield"
)
yield_df["Year"] = yield_df["Year"].str.replace("Yield-", "")

final_df = (
    area_df
    .merge(prod_df, on=id_cols + ["Year"], how="left")
    .merge(yield_df, on=id_cols + ["Year"], how="left")
)

final_df = final_df.dropna(subset=["Area", "Production", "Yield"], how="all")
```

## Adding Data Quality Flag
---
To improve data reliability, a **Data Quality Flag** is introduced to differentiate between:

* **Valid zero-production seasons** (genuinely no production),
* **Partial or incomplete reporting**, and
* **Structurally missing records**.

This enables tailored filtering depending on the analysis requirements.

The flags are categorized into three groups:

* **Zero Production:** All three — area, production, and yield — equal zero, indicating a valid zero-production season.
* **Good:** At least one of the area, production, or yield values is greater than zero, representing reliable data.
* **Insufficient:** Data points with missing (`NaN`) values or inconsistent entries that do not meet the above criteria.

```python
cols = ["area_mha", "production_mt", "yield"]

def data_quality_flag(row):
    values = row[cols].infer_objects(copy=False)

    # Case 1: all explicitly zero → valid no-production season
    if values.fillna(0).eq(0).all() and values.notna().all():
        return "ZERO_PRODUCTION"

    # Case 2: at least one meaningful value
    if (values > 0).any():
        return "GOOD"

    # Case 3: everything else (zeros + NaNs)
    return "INSUFFICIENT"

final_df["data_quality_flag"] = final_df.apply(data_quality_flag, axis=1)
```

## Validation of Aggregated Crops
---
The dataset includes an aggregated category **"Jute & Mesta"**, which should represent the sum of:

- Jute
- Mesta

To ensure data accuracy, a validation check compares:

Combined Production (Jute + Mesta)  
vs  
Reported Production (Jute & Mesta)

If the values match within a tolerance range, the record is considered consistent.

This validation helps detect **aggregation errors or inconsistencies in the dataset**.

```python
def validate_jute_mesta(
    df: pd.DataFrame,
    abs_tol: float = 1.0,
    rel_tol: float = 0.01,
    drop_matches: bool = True
):
    """
    Validates 'Jute & Mesta' production against
    sum of 'Jute' and 'Mesta'.

    Parameters
    ----------
    drop_matches : bool
        If True, removes matching rows.
        If False, keeps all rows.

    Returns
    -------
    df_result : pd.DataFrame
    stats : dict
    """

    required_cols = {'state', 'year', 'season', 'crop', 'production_mt'}
    missing = required_cols - set(df.columns)
    if missing:
        raise ValueError(f"Missing required columns: {missing}")

    keys = ['state', 'year', 'season']
    df = df.copy()

    component_df = df[df['crop'].isin(['Jute', 'Mesta'])]

    expected_df = (
        component_df
        .groupby(keys, as_index=False)['production_mt']
        .sum()
        .rename(columns={'production_mt': 'expected_prod'})
    )

    combined_df = df[df['crop'] == 'Jute & Mesta']

    check_df = combined_df.merge(expected_df, on=keys, how='left')

    difference = (check_df['production_mt'] - check_df['expected_prod']).abs()

    pct_error = np.where(
        check_df['expected_prod'].notna() & (check_df['expected_prod'] != 0),
        difference / check_df['expected_prod'],
        np.nan
    )

    has_component = check_df['expected_prod'].notna()

    is_match = (
        has_component &
        ((difference <= abs_tol) | (pct_error <= rel_tol))
    )

    # Stats
    stats = {
        "total_jm_rows": len(check_df),
        "matches": int(is_match.sum()),
        "non_matches": int((~is_match & has_component).sum()),
        "no_component_data": int((~has_component).sum())
    }

    if drop_matches:
        df_result = pd.concat([
            df[df['crop'] != 'Jute & Mesta'],
            check_df.loc[~is_match, df.columns]
        ], ignore_index=True)
    else:
        df_result = df.copy()

    return df_result, stats
```

## Adding Crop Category
---
A Crop Category column is introduced to group crops into broader categories based on their type and usage.

This categorization allows Aggregated analysis by crop type, Easier comparison across categories, and Simplified visualization of production patterns.

```python
category_map = {
    # Food Grains
    'food_grains' : [
        "Nutri Cereals", "Cereals", "Rice", "Maize", "Jowar", 
        "Bajra", "Wheat", "Small Millets", "Ragi", "Barley"
    ],
    # Pulses
    'pulses' : [
        "Other Pulses", "Urad", "Moong", "Tur", "Gram", "Lentil"
    ],
    # Oilseeds
    'oilseeds' : [
        "Groundnut", "Sunflower", "Sesamum", "Rapeseed & Mustard", 
        "Soybean", "Castorseed", "Linseed", "Nigerseed", "Safflower", "Guarseed"
    ],

    # Remaining / Other Crops
    "industrial_crops" : [
        "Cotton", "Sugarcane", "Tobacco", "Jute", "Mesta", "Sannhemp"
    ],
}
```

```python
crop_to_category = {
    crop: category
    for category, crops in category_map.items()
    for crop in crops
}

final_df["crop_category"] = final_df["crop"].map(crop_to_category)
```
